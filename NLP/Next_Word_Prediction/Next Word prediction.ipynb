#### Text preprocessing

```python
import re
```

```python
re_special_char = "[^a-z\s]"
```

```python
def text_preprocessing(review):
    review = str(review)
    #Normalizing the case 
    review = review.lower()
    #Removal of special characters
    review = re.sub(re_special_char,"", review)
    return review
```

```python
tweets["Tweets"]  = tweets["Tweets"].apply(text_preprocessing)
```
